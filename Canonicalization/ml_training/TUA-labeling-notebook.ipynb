# Labelling events!

This notebook will be for all of us to **label** our training events. The process will look something like:

1. Read each TUA and label its event ID

2. Write it down

3. Produce a file that contains the event IDs



```python
### IGNORE THIS PART ###
def tua_cleaner(tua):
    return tua.strip('c').strip('(').strip(')')
########################

reading_set = Table().read_table('TUA_reading_set.csv')
reading_set = reading_set.with_column("Cleaner TUA", reading_set.apply(tua_cleaner, "TUA"))
reading_set.show()
```

```python
reading_set.group('event_place').sort(1, descending = True).where('count', are.above(1)).column(1).sum()
```

What follows is a work flow that will allow us to quickly and thoroughly read through all of the TUAs and label the events they refer to.

The idea is that:
