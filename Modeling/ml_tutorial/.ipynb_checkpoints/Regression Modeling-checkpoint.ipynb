## Logistic regression

Unlike linear regression, which tries to predict something **continuous** using data (what's the height of a son based on his father's height? what is someone's weight if you know their height? etc.), logistic regression handles the **binary** prediction case. When we say "binary," we mean that the outcomes we're trying to predict, instead of being some number on a continuous scale, has 2 classes. 

Say you forget what dogs and cats look like (and the world only contains dogs and cats). You and your friend weigh some dogs and cats and she labels which ones are which. So now you have a dataset relating weight to animal. You're walking down the street one day and see an animal (which again can only be a dog or a cat). So you whip out your scale and weigh the animal. Theoretically based on its weight, you should be able to tell whether it's a dog or a cat.

That's a silly example, but it illustrates a whole *type* of prediction problem we often handle: classification. In this case, it's binary classification. Later on, we may see some techniques we can use for classification with multiple classes.

#### The Iris dataset

```python
iris = datasets.load_iris()
iris_sv = iris['data'][iris['target'] <= 1]
target = iris['target'][iris['target'] <= 1]
iris['DESCR'].split('\n')
```

![](https://www.math.umd.edu/~petersd/666/html/iris_with_labels.jpg)

This dataset is famous among learners and practitioners of classification for its clarity of purpose. For a given iris, based on the dimensions of the sepals and the dimensions of the petals, we should be able to tell what species of iris it is. In our case, we're only going to be looking at distinguishing between two species: `Setosa` and `Versicolour`.

```python
X = iris_sv[:, 0].reshape(-1, 1)
y = target
ax = sns.regplot(X, y)
ax.set(xlabel = 'Sepal length (cm)', ylabel = 'Class')
```

Here, we're looking at 1 variable, the sepal length, and the class (species). As we can see from above, fitting a linear regression model doesn't make a lot of sense here. Particularly, it doesn't make much sense for the line to extend below 0 or above 1 (what does a class <0 or >1 even mean?). Instead of fitting a line to our data, we'll work on fitting a curve that matches these constraints.

```python
ax2 = sns.regplot(X, y, logistic = True)
ax2.set(xlabel = 'Sepal length (cm)', ylabel = 'Class')
```

```python
log_reg_model = linear_model.LogisticRegression()
log_reg_model.fit(X, y)
metrics.confusion_matrix(log_reg_model.predict(X), y)
```

Above is what is called a *confusion matrix*. This is a way to visualize the predictions that a classifier makes.

![](http://rasbt.github.io/mlxtend/user_guide/evaluate/confusion_matrix_files/confusion_matrix_1.png)

When we talk about the *accuracy* of a classifier, what we mean is $\frac{TP + TN}{\text{all}}$. So the accuracy for our classifier, which uses `Sepal Length` to predict `Species`, is: $(39 + 45)/100 = 84\%$.

```python
metrics.accuracy_score(log_reg_model.predict(X), y)
```

Instead of just using one variable, let's try using all the data that we have to predict the classes! Unless a variable is *really, really* bad (basically uncorrelated and noisy), adding more variables should help us make predictions.

```python
X = iris_sv
log_reg_model.fit(X, y)
metrics.confusion_matrix(log_reg_model.predict(X), y)
```

Accuracy: 100%!

### Applying logistic regression to the canonicalized dataset

The canonicalized dataset is a little weird. Basically, all of our data is binary. In the `iris` example, our data was continuous with the exception of the thing we were trying to predict (often referred to as the "target"). Normally, having all binary data, both predictors and the target, is difficult to make sense of. Luckily, logistic regression doesn't care much about that, since it uses what are called *log-odds* to make predictions.

```python
canon_data = pd.read_csv('canonicalized_features_021118.csv').drop('Unnamed: 0', axis = 1).drop('TUA', axis = 1).drop('ids', axis = 1)

### you can ignore this part ###
# cols = list(canon_data)
# cols.insert(0, cols.pop(cols.index('ids')))
# canon_data = canon_data.loc[:, cols]

canon_data
```

```python
log_reg_model = linear_model.LogisticRegression()
X = canon_data.drop('4.01.01', axis = 1)
y = canon_data['4.01.01']
```

```python
log_reg_model.fit(X, y)
metrics.confusion_matrix(log_reg_model.predict(X), y)
```

#### The overfitting problem

Logistic regression is very "flexible," so to speak. Linear regression is very rigid, constrained by the fact that the data needs to look like a line generally. Logistic regression, as long as there is a binary target, can be very adaptable to the data. Which is why, when we "fit" it to some data, it can often do really, really well immediately. However, there's an issue. The fact that it's really accurate at predicting the data that we *trained it to predict* is sort of built into the fact that we trained it. A predictor/classifier is only good so long as it can predict **new, unseen** data reliably. So to account for this, we'll do what is called the *train-test split*. Instead of using all our data at once, we randomly select some of it to be held aside, representing unseen data. Then we fit our model on the rest of the data. Using the data we held aside, we can now simulate what it would look like if we let the model try to predict new, unseen data. This way of measuring accuracy and performance metrics is more honest and representative.

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.50)
log_reg_model.fit(X_train, y_train)
metrics.confusion_matrix(log_reg_model.predict(X_test), y_test)
```

And if we do this process many times, that is, if we hold some randomly chosen data aside, train a model, then predict on the data held aside, many times, it's called cross-validation. In cross-validation, we partition the data into $k$ parts, called folds. We choose 1 fold to be held aside, as the test set, and use the rest to train the model, as the training set. We can look at the performance across each of these folds individually, and if desired we can average them into what is referred to as the *CV score*.

```python
model_selection.cross_val_score(log_reg_model, X, y, cv = 3)
```
